Query 1

```python
# Uso df_steam
def juegos_lanzados_por_anio_y_porcentaje(developer):
    # Filtrar el dataframe por el developer especificado
    df_developer = df_steam[df_steam['developer'] == developer]
    
    # Agrupar por año y contar la cantidad de juegos lanzados
    juegos_por_anio = df_developer.groupby('release_date')['item_name'].count().reset_index()
    juegos_por_anio.columns = ['release_date', 'cantidad_juegos']
    
    # Calcular el porcentaje de juegos con precio 0.0 por año
    juegos_precio_cero_por_anio = df_developer[df_developer['price'] == 0.0].groupby('release_date')['item_name'].count().reset_index()
    juegos_precio_cero_por_anio.columns = ['release_date', 'cantidad_juegos_precio_cero']
    
    # Combinar los dos dataframes anteriores
    juegos_por_anio = juegos_por_anio.merge(juegos_precio_cero_por_anio, on='release_date', how='left')
    
    # Calcular el porcentaje de juegos con precio 0.0
    juegos_por_anio['porcentaje_juegos_precio_cero'] = (juegos_por_anio['cantidad_juegos_precio_cero'] / juegos_por_anio['cantidad_juegos']) * 100
    
    # Reemplazar los valores NaN por 0.0
    juegos_por_anio.fillna(0.0, inplace=True)
    
    # Crear un diccionario con la respuesta
    respuesta = {}
    for index, row in juegos_por_anio.iterrows():
        respuesta[int(row['release_date'])] = {
            'cantidad_juegos': row['cantidad_juegos'],
            'porcentaje_juegos_free': row['porcentaje_juegos_precio_cero']
        }
        
    # Devolver el diccionario de respuesta
    return respuesta
```

```python
# Ejemplo de uso de la función
respuesta = juegos_lanzados_por_anio_y_porcentaje("Valve")
print(respuesta)
```

```python
# Uso df_sub2
def obtener_informacion_juegos(user_id):
    # Filtrar el dataframe por el user_id especificado
    df_usuario = df_sub2[df_sub2['user_id'] == user_id]
    
    # Calcular el precio total de los juegos que posee
    precio_total = df_usuario['price'].sum()
    
    # Calcular el porcentaje de juegos en los que la columna "recommend" es verdadera
    porcentaje_recomendados = (df_usuario['recommend'].sum() / len(df_usuario)) * 100
    
    # Obtener la cantidad total de juegos que posee
    cantidad_juegos = len(df_usuario)
    
    # Crear un diccionario con la información
    informacion = {
        'user_id': user_id,
        'precio_total': precio_total,
        'porcentaje_recomendados': porcentaje_recomendados,
        'cantidad_juegos': cantidad_juegos
    }
    
    # Devolver el diccionario de información
    return informacion
```

Query 3

```python
# Uso df_sub1
def obtener_usuario_mas_horas_jugadas_por_genero(genero):
    # Filtrar el dataframe por el género especificado
    df_genero = df_sub1[df_sub1['genres'].apply(lambda x: genero in x)]
    
    # Obtener el usuario con más horas jugadas para el género dado
    usuario_mas_horas = df_genero.groupby('user_id')['playtime_forever'].sum().idxmax()
    
    # Filtrar nuevamente el dataframe por el usuario con más horas jugadas para el género dado
    df_usuario_mas_horas = df_genero[df_genero['user_id'] == usuario_mas_horas]
    
    # Obtener la acumulación de horas jugadas por año de lanzamiento
    acumulacion_horas_por_anio = df_usuario_mas_horas.groupby('release_date')['playtime_forever'].sum().reset_index()
    acumulacion_horas_por_anio.columns = ['Año', 'Horas']
    lista_acumulacion_horas = acumulacion_horas_por_anio.to_dict('records')
    
    # Crear el diccionario de respuesta
    respuesta = {
        "Usuario con más horas jugadas para " + genero: usuario_mas_horas,
        "Horas jugadas": lista_acumulacion_horas
    }
    
    # Devolver el diccionario de respuesta
    return respuesta
```

```python
# Ejemplo de uso
genero = 'Action'
respuesta = obtener_usuario_mas_horas_jugadas_por_genero(genero)
print(respuesta)
```

```python
genero = 'Racing'
respuesta = obtener_usuario_mas_horas_jugadas_por_genero(genero)
print(respuesta)
```

```python
# Uso df_sub2
def top_desarrolladores_recomendados_por_anio(anio):
    # Filtrar el dataframe por el año especificado y las condiciones de recomendación y análisis de sentimiento
    df_filtrado = df_sub2[(df_sub2['release_date'] == anio) & (df_sub2['recommend'] == True) & (df_sub2['sentiment_analysis'] == 2)]
    
    # Calcular el conteo de juegos recomendados por desarrollador
    conteo_juegos_por_desarrollador = df_filtrado['developer'].value_counts().reset_index()
    conteo_juegos_por_desarrollador.columns = ['Desarrollador', 'Cantidad']
    
    # Obtener el top 3 de desarrolladores con más juegos recomendados
    top_desarrolladores = conteo_juegos_por_desarrollador.head(3)
    
    # Crear una lista de diccionarios con el resultado
    resultado = []
    for index, row in top_desarrolladores.iterrows():
        puesto = "Puesto " + str(index + 1)
        desarrollador = row['Desarrollador']
        resultado.append({puesto: desarrollador})
    
    # Devolver el resultado
    return resultado
```

```python
# Ejemplo
anio = 2000
resultado = top_desarrolladores_recomendados_por_anio(anio)
print(resultado)
```

```python
anio = 2004
resultado = top_desarrolladores_recomendados_por_anio(anio)
print(resultado)
```

```python
# Uso df_sub2
def sumar_sentimientos_por_desarrollador(developer):
    # Filtrar el dataframe por el desarrollador especificado
    df_filtrado = df_sub2[df_sub2['developer'] == developer]
    
    # Sumar la cantidad de registros de sentiment_analysis como "Positivo" si es 2 y "Negativo" si es 0
    positivo = df_filtrado[df_filtrado['sentiment_analysis'] == 2].shape[0]
    negativo = df_filtrado[df_filtrado['sentiment_analysis'] == 0].shape[0]
    
    # Crear un diccionario con los resultados
    resultado = {
        developer: {
            'Negativo': negativo,
            'Positivo': positivo
        }
    }
    
    # Devolver el diccionario de resultados
    return resultado
```

```python
# Ejemplo de uso
developer = 'Valve'
resultado = sumar_sentimientos_por_desarrollador(developer)
print(resultado)
```
