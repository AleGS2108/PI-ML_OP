CARGA DEL ARCHIVO "OTUPUT_STEAM_GAMES.JSON"

```python
import pandas as pd
import json
```

```python
# Se carga el primer dataframe
data = []
with open('../output_steam_games.json', 'r') as file:
    for line in file:
        data.append(json.loads(line))
df_steam = pd.DataFrame(data)
```

```python
# Se eliminan todas aquellas filas que posee todos sus valores nulos en todas las columnas
df_steam=df_steam.dropna(how='all')
df_steam.info()
```

```python
# Se eliminan aquellas columnas innecesarias para las querys de la APIs y se reindexa
columns=['url','reviews_url','specs','publisher','early_access','tags']
df_steam=df_steam.drop(columns,axis=1).reset_index(drop=True)
```

Proceso de transformación y limpieza de columnas

```python
df_steam.head()
```

Columna "id"

```python
# Se verifican si hay filas con "id" nulo 
filtro = df_steam['id'].isna()
filas_filtradas = df_steam[filtro]
filas_filtradas
```

```python
# Se eliminan dichas filas
df_steam=df_steam.dropna(subset=['id']).reset_index(drop=True)
```

Columna "developer"

```python
# Se verifican si hay filas con "developer" nulo 
filtro = df_steam['developer'].isnull()
filas_filtradas = df_steam[filtro]
filas_filtradas
```

```python
# Se eliminan dichas filas
df_steam=df_steam.dropna(subset=['developer']).reset_index(drop=True)
```

Columna "genres"

```python
df_steam=df_steam.dropna(subset=['genres']).reset_index(drop=True)
df_steam
```

```python
# Se crea una columna auxiliar que muestra si incluye "Free to play" como genre
df_steam['incluye_free'] = df_steam.apply(lambda row: 'si' if 'Free to Play' in row['genres'] else 'no', axis=1)
```

```python
df_steam
```

```python
# En caso de que la columna "incluye_free" sea si, cambio el valor de price a 0.0
df_steam.loc[df_steam['incluye_free'] == 'si', 'price'] = 0.0
# Se elimina la etiqueta "Free to Play" de las listas de 'genres' y se eliminan las columnas auxiliares
df_steam.loc[df_steam['incluye_free'] == 'si', 'genres'] = df_steam.loc[df_steam['incluye_free'] == 'si', 'genres'].apply(lambda x: [genre for genre in x if genre != 'Free to Play'])
df_steam=df_steam.drop('incluye_free',axis=1)
```

```python
df_steam
```

Columna "price"

```python
# Se observa la columna "price" creando una columna auxiliar para ver si se puede pasar a tipo float
def check_float(value):
    try:
        float(value)
        return True
    except ValueError:
        return False
df_steam['is_float'] = df_steam['price'].apply(check_float)
```

```python
df_steam
```

```python
# Se observan la cantidad de cada tipo en esta columna auxiliar
tipo_counts=df_steam['is_float'].value_counts()
tipo_counts
```

```python
# Vemos aquellos valores de la columna "price" donde is_float es False
filtered_df = df_steam[df_steam['is_float']==False]
price_counts = filtered_df['price'].value_counts()
price_counts
```

```python
# Eliminar las filas que cumplen las condiciones
condiciones = (df_steam['price'] == 'Play Now') | (df_steam['price'] == 'Third-party') | (df_steam['price'] == 'Install Theme') | (df_steam['price'] == 'Install Now')
df_steam= df_steam.drop(df_steam[condiciones].index)
```

```python
filtered_df = df_steam[df_steam['is_float']==False]
price_counts = filtered_df['price'].value_counts()
price_counts
```

```python
# Cambiar los valores de la columna "price" a 0.0 si "is_float" es False
df_steam.loc[df_steam["is_float"] == False, "price"] = 0.0
# Eliminamos la columna auxiliar
df_steam = df_steam.drop('is_float', axis=1)
```

```python
df_steam
```

```python
df_steam=df_steam.dropna(subset=['price']).reset_index(drop=True)
```

```python
import ast
data=[]
with open('../australian_users_items.json','r',encoding='utf-8') as f:
    for line in f:
        data.append(ast.literal_eval(line))
df_user=pd.DataFrame(data)
```

```python
# Elimino las columnas inncesarias
df_user_final = df_user_final.drop(["items_count","user_url", "playtime_2weeks",'steam_id','item_name'], axis=1)
```

```python
# Revisamos si hay valores faltantes en alguna columna
filas_nulas_user_id = df_user_final[df_user_final['user_id'].isnull()].shape[0]
filas_nulas_item_id = df_user_final[df_user_final['item_id'].isnull()].shape[0]
filas_nulas_playtime_forever = df_user_final[df_user_final['playtime_forever'].isnull()].shape[0]
print("Número de filas con valores nulo en user_id:", filas_nulas_user_id)
print("Número de filas con valores nulo en item_id:", filas_nulas_item_id)
print("Número de filas con valores nulo en playtime_forever:", filas_nulas_playtime_forever)
```

```python
# Eliminamos las filas donde "playtime_forever" es nulo y reindexamos
df_user_final = df_user_final.dropna(subset=['playtime_forever']).reset_index(drop=True)
```

CARGA DEL ARCHIVO "AUSTRALIAN_USER_REVIEWS.JSON"

```python
with open('../australian_user_reviews.json', 'r', encoding='utf-8') as file:
    data = [eval(line) for line in file]
df_reviews = pd.DataFrame(data)
```

```python
user_reviews_explored = df_reviews.explode('reviews')
user_reviews_normalized = pd.json_normalize(user_reviews_explored['reviews'])
user_reviews_explored.reset_index(drop=True, inplace=True)
user_reviews_normalized.reset_index(drop=True, inplace=True)
df_reviews_final = pd.concat([user_reviews_explored.drop('reviews',axis=1), user_reviews_normalized],axis=1)
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
# Descargar los recursos necesarios para el análisis de sentimiento
nltk.download('vader_lexicon')
# Crear una instancia del analizador de sentimientos
sid = SentimentIntensityAnalyzer()
# Función para clasificar el sentimiento en la escala deseada
def get_sentiment_score(text):
    if isinstance(text, str) and text.strip() != '':
        sentiment = sid.polarity_scores(text)
        compound_score = sentiment['compound']
        if compound_score >= 0.05:
            return 2  # Positivo
        elif compound_score <= -0.05:
            return 0  # Negativo
        else:
            return 1  # Neutral
    else:
        return 1  # Valor por defecto si la columna "review" está vacía
# Aplicar el análisis de sentimiento a la columna "review" y crear la columna "sentiment_analysis"
df_reviews_final['sentiment_analysis'] = df_reviews_final['review'].apply(get_sentiment_score)
```

```python
# Se verifica el total de cada tipo en la columna "sentiment_analysis"
count_0 = df_reviews_final[df_reviews_final['sentiment_analysis'] == 0].shape[0]
count_1 = df_reviews_final[df_reviews_final['sentiment_analysis'] == 1].shape[0]
count_2 = df_reviews_final[df_reviews_final['sentiment_analysis'] == 2].shape[0]
print("Cantidad de filas con valor 0 en sentiment_analysis:", count_0)
print("Cantidad de filas con valor 1 en sentiment_analysis:", count_1)
print("Cantidad de filas con valor 2 en sentiment_analysis:", count_2)
```

```python
# Revisamos valores faltantes en las columnas
filas_nulas_user_id = df_reviews_final[df_reviews_final['user_id'].isnull()].shape[0]
filas_nulas_item_id = df_reviews_final[df_reviews_final['item_id'].isnull()].shape[0]
filas_nulas_recommend = df_reviews_final[df_reviews_final['recommend'].isnull()].shape[0]
print("Número de filas con valores nulo en user_id:", filas_nulas_user_id)
print("Número de filas con valores nulo en item_id:", filas_nulas_item_id)
print("Número de filas con valores nulo en recommend:", filas_nulas_recommend)
```

```python
df_reviews_final = df_reviews_final.dropna(subset=['item_id']).reset_index(drop=True)
```

```python
# Eliminamos las columnas innecesarias
df_reviews_final = df_reviews_final.drop(["funny","user_url", "posted", "last_edited", "helpful",'review'], axis=1)
df_reviews_final
```
